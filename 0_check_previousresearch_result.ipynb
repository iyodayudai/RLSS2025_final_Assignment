```python
NUM_CORES = 16
warnings.filterwarnings("ignore")
```

```python
agent1=tk.Agent("agent1",
                model=tm.ToyModel_SA.copy(),
                actor_network=tk.NN,
                critic_network=tk.NN,
                clip=0.1,
                lr_actor=0.0001,
                lr_critic=0.001,
                grad_updates=10,
                optimizer_actor=torch.optim.Adam,
                optimizer_critic=torch.optim.Adam,
                observables=['agent1' ,'Glc', 'Starch'],
                actions=["Amylase_e"],
                gamma=1,
                )

agents=[agent1]
```

```python
env_1=tk.Environment(name="Toy-Exoenzyme-Single-agents",
                    agents=agents,
                    dilution_rate=0.0001,
                    initial_condition={"Glc":100,"agent1":0.1,"Starch":10},
                    inlet_conditions={"Starch":10},
                    extracellular_reactions=[{"reaction":{
                    "Glc":10,
                    "Starch":-0.1,},
                    "kinetics": (tk.general_kinetic,("Starch","Amylase"))}],
                     dt=0.1,
                     number_of_batches=5000,
                     episodes_per_batch=int(NUM_CORES/2),)
```

```python
sim_1=tk.Simulation(name=env_1.name,
                  env=env_1,
                  save_dir="./MyResults/",
                  )
```

```python
import inspect
import spamdfba.toolkit as tk

print("rollout signature:", inspect.signature(tk.rollout))  # デバッグ用

# 元の rollout を退避
_orig_rollout = tk.rollout

# num_workers を省略可能にするラッパーを被せる
def _rollout_patched(env, num_workers=None, parallel_framework="native"):
    if num_workers is None:
        # 既存コードの意図に合わせて、エピソード並列数をデフォルトに
        num_workers = getattr(env, "episodes_per_batch", 1)
    # 元のrolloutは (env, num_workers) だけ（古い版）か、
    # もしくは (env, num_workers, parallel_framework) の場合があります。
    try:
        return _orig_rollout(env, num_workers, parallel_framework)  # 新しめの版
    except TypeError:
        return _orig_rollout(env, num_workers)                      # 古い版

# 差し替え
tk.rollout = _rollout_patched
```

```python
def _find_ex_rxn(env, sub="Starch", prod="Glc"):
    # Starch を負、Glc を正に持つ外因性反応を探す（なければ None）
    for i, exr in enumerate(env.extracellular_reactions):
        r = exr["reaction"]
        if sub in r and prod in r and r[sub] < 0 and r[prod] > 0:
            return i
    return None

def _kinetic_rate(env, ex_idx):
    if ex_idx is None: 
        return np.nan
    kin_fn, kin_args = env.extracellular_reactions[ex_idx]["kinetics"]
    vals = [env.state[env.species.index(m)] for m in kin_args]
    return float(kin_fn(*vals))

def evaluate_trained_policy(env, steps=None, reset=True, deterministic=True, plot=True):
    """
    学習済みポリシーで評価ロールアウト。探索ノイズは使わず平均値で行動します。
    - steps: None なら env.episode_length を使用
    - reset: True なら初期条件から、False なら現在状態から続きで評価
    - deterministic: True で俳優ネットの平均値を行動に採用
    返り値: dict（時系列を numpy 配列で格納）
    """
    steps = steps or getattr(env, "episode_length", 200)
    # 退避
    training_bak = getattr(env, "training", False)
    a_bak = [np.array(getattr(ag, "a", np.zeros(1)), dtype=float) for ag in env.agents]
    try:
        env.training = False
        if reset:
            env.reset()

        i_glc = env.species.index("Glc") if "Glc" in env.species else None
        i_sta = env.species.index("Starch") if "Starch" in env.species else None
        ex_idx = _find_ex_rxn(env, "Starch", "Glc")

        T, GLC, STA, ACT, GROW, INL_GLC, INL_STA, RATE = [], [], [], [], [], [], [], []

        for _ in range(steps):
            # 観測→行動（deterministic: 俳優ネットの出力そのまま）
            for ag in env.agents:
                # 観測ベクトル（あなたの実装では observables + 時刻 を入力）
                tfeat = getattr(env, "_t_in_episode", 0.0)
                obs = env.state[ag.observables]
                x = np.hstack([obs, tfeat]).astype(np.float32)
                with torch.no_grad():
                    act = ag.actor_network_(torch.tensor(x)).numpy()
                ag.a = np.array(act, dtype=float)

            # 記録（step前）
            T.append(getattr(env, "_t_in_episode", (T[-1] + env.dt) if T else 0.0))
            if i_glc is not None: 
                GLC.append(env.state[i_glc])
                INL_GLC.append(env.inlet_conditions[i_glc])
            else:
                GLC.append(np.nan); INL_GLC.append(np.nan)
            if i_sta is not None: 
                STA.append(env.state[i_sta])
                INL_STA.append(env.inlet_conditions[i_sta])
            else:
                STA.append(np.nan); INL_STA.append(np.nan)
            RATE.append(_kinetic_rate(env, ex_idx))
            ACT.append(float(env.agents[0].a[0]))  # 1変数（Amylase）の想定

            # 1ステップ進める
            s, r, a, sp = env.step()

            # 成長率（報酬）は r（各エージェント）を利用（ここでは agent1 を例示）
            GROW.append(float(r[0]))

        out = {
            "time": np.array(T),
            "glc": np.array(GLC),
            "starch": np.array(STA),
            "inlet_glc": np.array(INL_GLC),
            "inlet_starch": np.array(INL_STA),
            "amylase_action": np.array(ACT),
            "growth": np.array(GROW),
            "starch_to_glc_rate": np.array(RATE),
        }

        if plot:
            fig, axs = plt.subplots(4, 1, figsize=(11, 10), sharex=True)
            axs[0].plot(out["time"], out["starch"], label="C[Starch]")
            axs[0].plot(out["time"], out["inlet_starch"], "--", label="C_in[Starch]")
            axs[0].set_ylabel("Starch"); axs[0].legend()

            axs[1].plot(out["time"], out["glc"], label="C[Glc]")
            axs[1].plot(out["time"], out["inlet_glc"], "--", label="C_in[Glc]")
            axs[1].set_ylabel("Glc"); axs[1].legend()

            axs[2].plot(out["time"], out["amylase_action"])
            axs[2].set_ylabel("Amylase (policy)")

            axs[3].plot(out["time"], out["growth"], label="reward/growth")
            axs[3].plot(out["time"], out["starch_to_glc_rate"], label="rate(Starch→Glc)")
            axs[3].set_ylabel("rate / growth"); axs[3].set_xlabel("time"); axs[3].legend()

            plt.tight_layout(); plt.show()

        return out

    finally:
        # 復元
        env.training = training_bak
        for ag, a0 in zip(env.agents, a_bak):
            ag.a = a0
```

```python
#学習後のポリシーを評価
import matplotlib.pyplot as plt 
def get_agent_obs_info(env, agent_idx=0):
    """エージェントが見ている観測の '名前' と 'env.state インデックス' を返す"""
    ag = env.agents[agent_idx]
    obs_idx = ag.observables                  # すでに index 化済み
    obs_names = [env.species[i] for i in obs_idx]
    return obs_names, obs_idx

@torch.no_grad()
def policy_eval_batch(env, agent_idx, obs_batch_with_t):
    """obs(＋t)のバッチに対して、平均出力を返す（numpy）"""
    ag = env.agents[agent_idx]
    x = torch.tensor(obs_batch_with_t, dtype=torch.float32)
    mu = ag.actor_network_(x).cpu().numpy()   # 行動は連続（例：Amylase 分泌）
    return mu

def plot_policy_1d(env, agent_idx=0, vary="Glc", fixed=None, t_fixed=0.0, 
                   x_min=0.0, x_max=50.0, n=200):
    """
    1変数 vary を動かし、他の観測は fixed で固定して方策（Amylase）を曲線で表示。
    fixed は dict で {観測名: 値} を指定（例：{"agent1":0.1, "Starch":10}）
    """
    obs_names, obs_idx = get_agent_obs_info(env, agent_idx)
    if fixed is None: fixed = {}
    if vary not in obs_names:
        raise ValueError(f"'{vary}' は観測に含まれていません。obs={obs_names}")

    # 固定ベクトルを用意（観測次元）
    base = np.zeros(len(obs_names), dtype=np.float32)
    for k, v in fixed.items():
        if k not in obs_names:
            raise ValueError(f"fixed に未知のキー: {k} （観測は {obs_names}）")
        base[obs_names.index(k)] = float(v)

    xs = np.linspace(x_min, x_max, n)
    batch = []
    for val in xs:
        vec = base.copy()
        vec[obs_names.index(vary)] = float(val)
        batch.append(np.hstack([vec, t_fixed]))   # 末尾に時刻特徴を付与（あなたの実装に合わせる）
    batch = np.asarray(batch, dtype=np.float32)

    mu = policy_eval_batch(env, agent_idx, batch)   # shape: (n, num_actions)
    if mu.ndim == 2 and mu.shape[1] == 1:
        mu = mu[:,0]

    plt.figure(figsize=(7,4))
    plt.plot(xs, mu)
    plt.xlabel(vary)
    plt.ylabel("policy output (Amylase)")
    ttl_fixed = ", ".join([f"{k}={fixed[k]}" for k in fixed]) if fixed else "none"
    plt.title(f"Policy 1D: vary {vary} @ t={t_fixed}, fixed [{ttl_fixed}]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    return xs, mu

def plot_policy_2d(env, agent_idx=0, vary_x="Glc", vary_y="Starch", fixed=None, t_fixed=0.0,
                   x_min=0.0, x_max=50.0, y_min=0.0, y_max=50.0, nx=80, ny=80):
    """
    2変数 (vary_x, vary_y) を動かし、他の観測を fixed で固定して方策をヒートマップ表示。
    """
    obs_names, obs_idx = get_agent_obs_info(env, agent_idx)
    if fixed is None: fixed = {}
    for name in [vary_x, vary_y]:
        if name not in obs_names:
            raise ValueError(f"'{name}' は観測に含まれていません。obs={obs_names}")

    base = np.zeros(len(obs_names), dtype=np.float32)
    for k, v in fixed.items():
        if k not in obs_names:
            raise ValueError(f"fixed に未知のキー: {k} （観測は {obs_names}）")
        base[obs_names.index(k)] = float(v)

    xs = np.linspace(x_min, x_max, nx)
    ys = np.linspace(y_min, y_max, ny)
    grid = []
    for y in ys:
        for x in xs:
            v = base.copy()
            v[obs_names.index(vary_x)] = float(x)
            v[obs_names.index(vary_y)] = float(y)
            grid.append(np.hstack([v, t_fixed]))      # 観測 + time
    grid = np.asarray(grid, dtype=np.float32)
    mu = policy_eval_batch(env, agent_idx, grid)      # (nx*ny, num_actions)
    if mu.ndim == 2 and mu.shape[1] == 1:
        mu = mu[:,0]
    Z = mu.reshape(ny, nx)

    plt.figure(figsize=(7,5))
    im = plt.imshow(Z, origin="lower", extent=[x_min,x_max,y_min,y_max], aspect="auto")
    plt.colorbar(im, label="policy output (Amylase)")
    plt.xlabel(vary_x); plt.ylabel(vary_y)
    ttl_fixed = ", ".join([f"{k}={fixed[k]}" for k in fixed]) if fixed else "none"
    plt.title(f"Policy 2D @ t={t_fixed}, fixed [{ttl_fixed}]")
    plt.tight_layout()
    plt.show()
    return xs, ys, Z
```
